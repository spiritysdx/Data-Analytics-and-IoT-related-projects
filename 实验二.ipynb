```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing

plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 

data1 = pd.read_csv('2.csv', encoding='gbk')
```

```python
data1 = data1.set_index(['样本序号'])
data1
```

```python
data2 = preprocessing.scale(data1)
data2
```

```python
# 计算相关系数矩阵
data3 = np.around(np.corrcoef(data2.T),decimals=3)
data3
```

```python
# 计算特征值，特征向量
featValue, featVec=  np.linalg.eig(data3.T)
featValue, featVec
```

```python
#求特征值的贡献度
gx = featValue/np.sum(featValue)
gx
```

```python
#选出主成分 k = 3 三个特征值贡献率相加>85
k = [0, 1, 2]
k
```

```python
#选出主成分对应的特征向量矩阵
selectVec = np.matrix(featVec.T[k]).T
selectVe=selectVec*(-1)
selectVec
```

```python
df1=data2*np.array(pd.DataFrame(selectVec).loc[:,0])
tf1=df1*gx[0]/100
pd.DataFrame(tf1,index=data1.index, columns=data1.columns).sort_values(["x1：人口密度(人/km2)"])
```
